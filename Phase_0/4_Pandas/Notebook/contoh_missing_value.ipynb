```python
missing_values = ["na","--"]
data = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/property_data.csv', na_values = missing_values)
```

```python
cnt=0
for row in data['OWN_OCCUPIED']:
    try:
        int(row)
        data.loc[cnt, 'OWN_OCCUPIED']=np.nan
    except ValueError:
        pass
    cnt+=1
# mengganti nilai yang berbeda sendiri pada 1 baris
# seperti baris seharusnya berisi huruf malah ada angka
```
